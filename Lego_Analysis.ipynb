```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/lego_sets.csv')
parent_theme = pd.read_csv('/content/sample_data/parent_themes.csv')
```

```python
df.head(5)
```

```python
parent_theme.head(10)
```

```python
merged = df.merge(parent_theme, left_on = 'parent_theme', right_on = 'name')
```

```python
merged.head(5)
```

```python
merged = merged.drop(columns = 'name_y')

merged.head(5)
```

```python
licensed = merged[merged['is_licensed'] == True]
```

```python
licensed.shape
```

```python
licensed = licensed.dropna(subset = ['set_num'])
```

```python
licensed.shape
```

```python
licensed.head(5)
```

```python
licensed_star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
```

```python
licensed_star_wars.shape
```

```python
the_force = int((licensed_star_wars.shape[0]*100)/licensed.shape[0])
```

```python
the_force
```

```python
licensed['is_star_wars'] = False
```

```python
licensed.head(5)
```

```python
licensed.loc[licensed['parent_theme'] == 'Star Wars', 'is_star_wars'] = True
```

```python
licensed_star_wars = licensed.loc[licensed['is_star_wars'] == True]
```

```python
licensed_star_wars_count = licensed_star_wars.groupby(['year']).count()['is_star_wars']
```

```python
licensed_star_wars_count
```

```python
licensed_total_count = licensed.groupby('year').count()['is_licensed']
```

```python
licensed_total_count
```

```python
list_licensed_star_wars_count = list(licensed_star_wars_count)
```

```python
list_licensed_total_count = list(licensed_total_count)
```

```python
licensed_star_wars_percent = []
```

```python
for index, value in enumerate(list_licensed_star_wars_count):
  percent = (value*100)/list_licensed_total_count[index]
  licensed_star_wars_percent.append(percent)
```

```python
licensed_star_wars_percent
```

```python
licensed_sorted = licensed.sort_values('year')
```

```python
licensed_sorted['count'] = 1
```

```python
summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
```

```python
summed_df
```

```python
new_df = summed_df.sort_values('count', ascending = False).drop_duplicates(['year'])
```
